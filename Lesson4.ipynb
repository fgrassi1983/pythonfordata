A quick recap on what we have seen so far:

### Variable types

We have learned about different variable types such as: 

* **int**    integers, e.g. 9 .
* **float**    numbers with decimals, e.g. 3.14.
* **str**    text, e.g. "hello".
* **list**    ordered containers where every item can be accessed by its index, e.g. mylist = [1, 4, "bla"], mylist[1] ---> 4. New items can be added with the .append method, e.g. mylist.append("hello"). 
* **dict**    unordered containers where every item can be accessed by its name, e.g. mydict = {"firstname":"Robert", "height":175}, mydict["userid"] ---> 123. New items can be added by declaration: mydict["weight"] = 70. 
* **tuple**    ordered containers like lists, e.g. (1, 4, "bla"), but immutable.

### Loops

**for** loops are used to iterate through any iterable object (e.g. lists, dictionaries, range(x,y)), and to repeat the same block of code at every cycle.

Example:

    for i in range(10):
        print(i)

**while** loops are used to repeat the same block of code as long as the condition specified in the while statement is true.

    guessthenumber = 9
    
    userguess = 0
    
    while userguess != guessthenumber:
        userguess = input("Guess")

### if ... elif ... else

**if** statements are used to run code only when a given condition is true.

    if x > 0:
        print("x is positive")

**elif** optionally follow after if statements, allowing for handling of multiple conditions.

    if x > 0:
        print("x is positive")
    elif x == 0:
        print("x is 0")

**else** statements, also optional, may go at the end of an if statement, as a catch all if all previous conditions were false. Note that else statements do not require a condition to be specified.

    if x > 0:
        print("x is positive")
    elif x == 0:
        print("x is 0")
    else:
        print("x is negative")

### Functions

Functions allow us to store a set of instructions which can then be used with the function's name. Functions can take any number of arguments, including zero. Example:

    def square(x):
        return x * x
    
    first10squared = []
    
    for i in range(1,11):
        first10squared.append(square(x))

### Lambda expressions

Lambda expressions are one line functions, which can be both named (i.e. assigned to a name, like variables) or anonymous (e.g. inside list(map()) constructs).

Named lambda:

    square = lambda x: x*x
    
    first10squared = []
    
    for i in range(1,11):
        isquared = square(i)
        first10squared.append(isquared)
        
Anonymous function:

    first10squared = list(map(lambda x: x * x, range(1,11)))

### List comprehensions

List comprehensions are ways to quickly populate a list:

    first10squared = [x*x for x in range(1,11)]

Usually, list comprehensions are faster than the other methods we have seen.

### Libraries

Libraries are code, written and tested by someone else, that can be imported to add functions to our code without having to implement them manually. E.g.

    import numpy as np
    
    studentage = [21,20,21,22,21,24,19,27,20]
    
    meanage = np.mean(studentage)
    
    medianage = np.median(studentage)

# Classes

While it is unlikely that you will have to use classes, they are nonetheless a very important topic in Python, and you should at least know what they are.

A **class** in Python is essentially a blueprint for a type of object. Suppose for example that we were making a racing video game, and we wanted to include different cars. We could use classes to define a generic car object as shown below:

```python
class car:
    """This is our car class."""
    def __init__(self, brand, name, color, mileage, maxspeed, pricenew):
        """And this is our init function, which initializes an instance of a car object."""
        self.brand = brand # Maker, e.g. Ferrari.
        self.name = name # Name, e.g. Testarossa.
        self.color = color # Color, e.g. red.
        self.mileage = mileage # Current kilometers, e.g. 0.
        self.maxspeed = maxspeed # Max speed, e.g. 88 mph.
        self.pricenew = pricenew # Price when new.
```

We have defined a class of objects called *car*, which has a name, a color, and a maxspeed. The strange looking function called **\_\_init\_\_** is a special method, which is used to **initialize** an object. It takes **self** as the first argument, and any number of optional arguments (e.g. name, color, maxspeed). Inside the **\_\_init\_\_** function, we declare as many variables (self.variablename = variablename) as the ones that follow **self**.

Now we can use the *car* class to define different types of car:

```python
class car:
    
    def __init__(self, brand, name, color, mileage, maxspeed, pricenew):
        self.brand = brand
        self.name = name
        self.color = color
        self.maxspeed = maxspeed
        self.mileage = mileage
        self.pricenew = pricenew
        self.pricecurrent = pricenew
    
    def priceupdate(self):
        """Just a simple function to make the price go down as milage increases.""" 
        self.pricecurrent = round(self.pricenew * (self.pricenew / (self.pricenew + self.mileage / 10)),2)

    def drive(self, speed, time):
        """The drive function will take in speed and time as input, 
        and it will return the distance travelled. It will also update 
        the car's km value, and its current price using the priceupdate function."""
        if speed > self.maxspeed: # Obviously cannot exceed the car's maximum's speed.
            speed = self.maxspeed
        distance = speed * time # Calculating distance.
        self.mileage = self.mileage + distance # Updating total km.
        self.priceupdate() # Updating the car's price.
        return distance

# Declaring the cars again.

niva = car("Lada","Niva","white",0,120,15000)
delorean = car("DeLorean","DMC","steel",0,177,40000)
veneno = car("Lamborghini","Veneno","black",0,356,3*10**6)
```

Pandas is a library for handling moderate amounts (i.e. up to a few million rows) of structured, tabular data. Should you ever need to work with larger amounts of data, i.e. enter Big Data territory, then perhaps Spark would be a better suited tool.

Pandas can read many different types of data, including csv, excel, and json.

In code, pandas is commonly shortened to **pd**.

```python
import pandas as pd
import numpy as np
```

There are two very important objects in Pandas: the **Series** and the **DataFrame**.

A pandas Series can be thought of as a vector, or a single column of values. It is in many ways similar to a list: for example, its individual items can be accessed through their index, and it is iterable:

```python
myseries = pd.Series([11,22,33])
```

```python
myseries[0]
```

```python
for i in myseries:
    print(i)
```

Let us now look at DataFrames. DataFrames are essentially a collection of Series, which together form a table. Let us see how to define a DataFrame from scratch:

```python
row0 = [np.random.randint(-10,-5) for i in range(3)] # Generating a list with three random numbers.
row1 = [np.random.randint(0,5) for i in range(3)] # And another one.
row2 = [np.random.randint(5,10) for i in range(3)] # As above.
row3 = [np.random.randint(-5,0) for i in range(3)] # So below.

allrows = [row0,row1,row2,row3]

for i in allrows:
    print(i)
```

```python
myfirstdf = pd.DataFrame(allrows,columns=["columna","columnb","columnc"])
myfirstdf
```

As you can see, we have four rows, because we passed a list with four items (allrows), and three columns, because each row had three items.

DataFrame objects can also be modified: suppose we wanted to add a fourth column, called columnd. We could do so using the following syntax:

```python
myfirstdf["columnd"] = [np.random.normal() for i in range(4)]
myfirstdf
```

By using the syntax *dataframename["columnname"] = something*, we created a new column called *columnname*, and set its contents to *something*.

We can also create new columns using existing ones, for example:

```python
myfirstdf["columne"] = myfirstdf["columna"] + myfirstdf["columnb"]
myfirstdf
```

Naturally, columns can also contain non-numeric values:

```python
myfirstdf["columnf"] = ["winter","spring","summer","autumn"]
myfirstdf
```

It may also happen that we need to use data from different DataFrames - if you have any experience with SQL, you will be very familiar with this situation. For example, suppose we wanted to see which cities are generating the most revenue. Revenue information is stored in the *sales* DataFrame, but cities are only found in the *customers* DataFrame. Luckily, our data engineers had the foresight of including a *userid* column in our *sales* DataFrame, which can be used to join *sales* with *customers*. This is done using the **merge()** method:

```python
merged = sales.merge(customers,
                     how="left",
                     left_on="userid",
                     right_on="id")

merged.head()
```

Let us examine the syntax. We called the **merge()** method on the *sales* DataFrame. The first argument we passed it was the DataFrame we wished to join it with, in this case, *customers*. The second argument is **how**, and it can most commonly be *left*, *inner*, or *right*. Do not worry too much about this for now, we will see how these options differ in the next lesson. The last two arguments, *left_on* and *right_on* are used to choose which columns we want to use to match entries from the left (*sales*) and right (*customers*) DataFrames. In this case, we used *userid* from *sales*, and *id* from *customers*, because we want to map our sales data with our customers data using their ids.

It really is not half as complicated as it sounds. It just takes some practice :)

Now, let us calculate the profit:

```python
merged["profit"] = (merged.pricesold - merged.pricebought) * merged.unitssold
```

```python
merged.head()
```

Because we want to see which city generated the most profits, we are going to have to aggregate by city. We can do this using the **groupby()** method, followed by **agg()** and the name of a function to calculate whatever we are interested in, in this case, **sum**:

```python
grouped = merged.groupby(by="city").agg(["sum","mean","median"])
grouped
```

To make visualization easier, we can finally use the **sort_values()** method:

```python
grouped.sort_values(by="profit",ascending=False)
```
